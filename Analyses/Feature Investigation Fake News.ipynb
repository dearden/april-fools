```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from pandas.plotting import scatter_matrix
```

```python
feats = pd.read_csv("../System/fake_news_feats_train.csv")
af = feats.loc[feats['class'] == 1]
naf = feats.loc[feats['class'] == 0]

top10_names = ['lexical_diversity', 'preposition', 'time_related', 'spelling_errors', 'numbers'
              ,'avg_sentence_len', 'usas_ambiguity', 'head_punctuation', 'readability', 'claws_ambiguity', 'class']
top_10 = feats.loc[:, top10_names]
af_top_10 = top_10.loc[feats['class'] == 1]
naf_top_10 = top_10.loc[feats['class'] == 0]
```

```python
def plot_feature_density(feat_name):
    feat = feats.loc[:, feat_name]
    af_feat = af.loc[:, feat_name]
    naf_feat = naf.loc[:, feat_name]
    
    fig, ax = plt.subplots()
    af_feat.plot.density(label='AF', ax=ax, figsize=(10,5), color='aqua')
    naf_feat.plot.density(label='NAF', ax=ax, color='coral')
    plt.legend()
    ax.set_xlabel(feat_name)
    plt.show()
```

```python
def plot_feature_line(feat_name):
    fig, ax = plt.subplots()
    
    feat = feats.loc[:, feat_name]
    af_feat = af.loc[:, feat_name]
    naf_feat = naf.loc[:, feat_name]
    
    af_feat.plot(color='aqua', ax=ax, figsize=(10,5))
    naf_feat.plot(color='coral', ax=ax)

    ax.axhline(af_feat.mean(), color='green')
    ax.axhline(naf_feat.mean(), color='crimson')
    ax.axhline(feat.mean(), color='purple')

    plt.show()
```

Lexical Diversity

```python
plot_feature_line("lexical_diversity")
plot_feature_density("lexical_diversity")
```

Preposition

```python
plot_feature_line("preposition")
plot_feature_density("preposition")
```
