```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

Read in the data

```python
feats = pd.read_csv("../System/feats_train.csv")
feats.index = list(range(1, len(feats) + 1))
feats.head()
```

Get the Predictions and add them to the data.

```python
with open("../System/feats_predictions.txt") as file:
    predictions = [int(line.strip()) for line in file]
    
feats['prediction'] = predictions
```

Look at errors (where predictions were different from class)

```python
errors = feats.loc[feats['class'] != feats['prediction']]

af_errors = errors.loc[errors['class'] == 1]
naf_errors = errors.loc[errors['class'] == 0]

corrects = feats.loc[feats['class'] == feats['prediction']]

af_corrects = corrects.loc[corrects['class'] == 1]
naf_corrects = corrects.loc[corrects['class'] == 0]

print("Num Articles: {}".format(feats.shape[0]))
print("Num Errors: {}".format(errors.shape[0]))
print("AF Errors: {}".format(af_errors.shape[0]))
print("NAF Errors: {}".format(naf_errors.shape[0]))
```

Read in metadata and list sites that errors were made on.

```python
column_names = ["Index", "URL", "Headline", "Year", "Category", "April Fool", "Set"]
meta = pd.read_csv("../Corpus Construction/meta.csv", delimiter=";", header=None, names=column_names)
meta.set_index('Index', inplace=True)
meta.head()
```

```python
error_sites = meta.loc[meta.index.isin(errors.index)]
error_sites.head()
```

```python
error_sites.to_csv("error_sites.csv")
```

```python
error_af_sites = meta.loc[meta.index.isin(af_errors.index)]
error_naf_sites = meta.loc[meta.index.isin(naf_errors.index)]
```

```python
from analysis import get_site
fig, ax = plt.subplots()
ax.get_xaxis().set_visible(False)

error_simplified_sites = error_sites.groupby(error_sites['URL'].apply(get_site), sort=False).size()
error_simplified_sites = error_simplified_sites.sort_values(ascending=False)
error_simplified_sites.plot(ax=ax)

error_af_simplified_sites = error_af_sites.groupby(error_af_sites['URL'].apply(get_site), sort=False).size()
error_af_simplified_sites = error_af_simplified_sites.sort_values(ascending=False)
error_af_simplified_sites.plot(ax=ax)

error_naf_simplified_sites = error_naf_sites.groupby(error_naf_sites['URL'].apply(get_site), sort=False).size()
error_naf_simplified_sites = error_naf_simplified_sites.sort_values(ascending=False)
error_naf_simplified_sites.plot(ax=ax)

plt.show()
```

```python
a = error_af_simplified_sites.value_counts().sort_index()
b = error_naf_simplified_sites.value_counts().sort_index()
c = error_simplified_sites.value_counts().sort_index()

fig, ax = plt.subplots()
# Need to try and set bottom axis to article index of AF.

ax = a.plot(ax=ax, label='AF')
ax = b.plot(ax=ax, label='NAF')
ax = c.plot(ax=ax, label='all')
# ax.set_xlim(xmin=0)
ax.set_ylim(ymin=0)
ax.set_ylabel("Websites in bin")
ax.set_xlabel("Number of site occurences")

plt.legend()

plt.show()
```

```python
mean_dif = pd.DataFrame(af_errors.mean(), columns=["Error Mean"])
mean_dif["Correct Mean"] = af_corrects.mean()
mean_dif = mean_dif.drop(['class', 'prediction'])
mean_dif['Difference'] = mean_dif['Error Mean'] - mean_dif['Correct Mean']
mean_dif.sort_values('Difference').head()
```

```python
std_dif = pd.DataFrame(af_errors.std(), columns=["Error Std"])
std_dif["Correct Std"] = af_corrects.std()
std_dif = std_dif.drop(['class', 'prediction'])
std_dif['Difference'] = std_dif['Error Std'] - std_dif['Correct Std']
std_dif.sort_values('Difference').head()
```
