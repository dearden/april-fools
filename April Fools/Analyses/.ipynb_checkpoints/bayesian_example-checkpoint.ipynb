# Parameter Tuning for IMDB Movie Review Classification

```python
#Simple examples of 3 different parameter tuning approaches
```

```python
#standard libraries
import os, re
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy
import pickle
%pylab inline

#ML libraries
from sklearn.cross_validation import train_test_split
from sklearn.feature_extraction.text import CountVectorizer, TfidfTransformer
from sklearn.feature_extraction.text import *
from sklearn.metrics import *
from sklearn.utils import shuffle
from sklearn import metrics
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import RandomizedSearchCV
from sklearn.model_selection import cross_validate
from sklearn.ensemble import RandomForestClassifier


#BO libraries
import GPy
import GPyOpt
rs=1234
```

```python
#data preparation and loading
np.random.seed(rs)
random.seed(rs)
text=[]
clas = []
classname = ["pos", "neg"]
#load training examples
for item in classname:
    for file in os.listdir("../../Data/aclImdb/test/" +item ):
            filename = "../../Data/aclImdb/test/" +item + "//"+file
            fl = open(filename, "r", encoding="utf8").read()
            fl = re.sub("\n", " ", fl)
            text.append(fl)
            clas.append(item)
#load testing examples
for item in classname:
    for file in os.listdir("../../Data/aclImdb/train/" +item ):
            filename = "../../Data/aclImdb/train/" +item + "//"+file
            fl = open(filename, "r", encoding="utf8").read()
            fl = re.sub("\n", " ", fl)
            text.append(fl)
            clas.append(item)
#store in dataframe
dataframe = pd.DataFrame(clas, columns=['class'])
dataframe["text"] = text
dataframe = shuffle(dataframe)
dataframe = dataframe.sample(frac=1).reset_index(drop=True)
print("We have "+str(len(text))+" classified examples")

#choose 5000 to be fixed training set
#leave remaing 45000 to be test set/represent the population
Y = dataframe["class"].tolist()
X = dataframe["text"].tolist()
X_train_fixed=X[:1000]
Y_train_fixed=Y[:1000]
X_pop=X[1000:]
Y_pop=Y[1000:]
#prepare for training
count_vect = CountVectorizer(min_df = 10, ngram_range = (1, 3),stop_words="english",analyzer='word',max_features=500)
X_train_counts = count_vect.fit_transform(X_train_fixed)
tfidf_transformer = TfidfTransformer()
X_train_fixed_tfidf = tfidf_transformer.fit_transform(X_train_counts)
X_pop_counts=count_vect.transform(X_pop)
X_pop_tfidf=tfidf_transformer.transform(X_pop_counts)
```

# Simple Grid Search

```python
#initialize classifier

#put own classifier here if you want!

rf=RandomForestClassifier(random_state=1234,n_estimators=1000,n_jobs=2)

#define grid
parameters = {'max_features':[0.1,0.9], 'max_depth':[1,3],'min_samples_split':[0.1,0.3],'min_samples_leaf':[0.1,0.3]}

#perform grid search with 5-fold CV

clf = GridSearchCV(rf, parameters, cv=5,n_jobs=10)
clf.fit(X_train_fixed_tfidf,Y_train_fixed)
```

```python
#return model chosen by grid search and its score
print(clf.best_estimator_)
print("giving accuracy of "+str(clf.best_score_))
```

# Suspiciously Good Random Search

```python
#initialize classifier

#put own classifier here if you want!

rf=RandomForestClassifier(random_state=1234,n_estimators=1000,n_jobs=2)

#define ranges for parameters
#any scipy.stats distributions can be used here (e.g if want to sample from log domain for regularization)
#or just a list for discrete variables

param_distributions={'max_features': scipy.stats.uniform(0,1), 'max_depth': [1,2,3,4,5],
  'min_samples_split': scipy.stats.uniform(0,0.5), 'min_samples_leaf':scipy.stats.uniform(0,0.5)}

#perform random search with 5-fold CV
#n_iter is number of random parameter choices to try 

clf_random = RandomizedSearchCV(rf, param_distributions,n_iter=81, cv=5,n_jobs=4)
clf_random.fit(X_train_fixed_tfidf,Y_train_fixed)
```

```python
#return model chosen by random search and its score
print(clf_random.best_estimator_)
print("giving accuracy of "+str(clf_random.best_score_))
```

```python
#return score of successively evaluated points
scores=clf_random.cv_results_["mean_test_score"]
#and plot the best previously observed solution at each iteration
plt.plot([x for x in range(1,len(scores))],[np.max(scores[:i]) for i in range(1,len(scores))])
plt.xlabel("Number of Evaluations")
plt.ylabel("Accuracy")
plt.axhline(y=.68, linewidth=2, color = 'red')
```

```python
#calc current best observed for each run of random_search
#load in some precomputed results for multiple runs of random-search
random_search_50=pickle.load( open( "random_search_50.p", "rb" )) 
#calc the best observed performances for each run at each eval
random_search_50_performance=np.array([[np.max(random_search_50[j,:i]) for i in range(1,shape(random_search_50)[1])] for j in range(0,shape(random_search_50)[0])])
```

```python
#plot indiviudal solution trajectories
evaluations=[x for x in range(1,len(random_search_50[0,:]))]
for j in range(0,shape(random_search_50)[0]):
    plt.plot(evaluations,random_search_50_performance[j,:])
plt.xlabel("Evaluation")
plt.ylabel("Accuracy")
plt.title("Best Observed Solution For 50 runs of Random Search")
plt.ylim([0.48,0.84])
```

```python
#plot average performance
means=np.mean(random_search_50_performance,axis=0)
error=np.sqrt(np.var(random_search_50_performance,axis=0))
plt.errorbar(evaluations, means, yerr=error, fmt='-o')
plt.xlabel("Number of Evaluations")
plt.ylabel("Accuracy")
#compare with solution from grid search
plt.axhline(y=.68, linewidth=2, color = 'red')
plt.ylim([0.48,0.82])
```

# Bayesian Optimization

```python
#set up domain of parameters
#specify if continous and give a range
# or if discrete give a list of possible values
domain=[{'name': 'max_features',      'type': 'continuous', 'domain': (0.00001,1)},
    {'name': 'max_depth', 'type':'discrete', 'domain':tuple(range(2,6))},
    {'name': 'min_samples_split',  'type': 'continuous', 'domain': (0.00000001,0.5)},
    {'name': 'min_samples_leaf',  'type': 'continuous', 'domain': (0.00000001,0.5)}] 
```

```python
#set up the function to optimize: The performance of our ML model
def fit_RF(x):
    x = np.atleast_2d(x)
    fs = np.zeros((x.shape[0],1))
    for i in range(x.shape[0]):
        clf=RandomForestClassifier(random_state=1234,n_estimators=1000,max_features=x[i,0],max_depth=int(x[i,1]),min_samples_split=x[i,2],min_samples_leaf=x[i,3],n_jobs=2)
        fs[i]=-np.mean(cross_validate(clf, X_train_fixed_tfidf, Y_train_fixed, cv=5,n_jobs=2)['test_score'])
    return fs
```

```python
#set up and initialize BO model
opt = GPyOpt.methods.BayesianOptimization(f =fit_RF,  # function to optimize       
                                          domain = domain,         # box-constrains of the problem
                                          acquisition_type ='LCB',
                                          initial_design_type="random",
                                          initial_design_numdata=15,
                                          kernel=GPy.kern.Matern52(len(domain)) #type of Gaussian Process most appropriate for ML models
                                         )   
```

```python
#perform max_iter evaluations
opt.run_optimization(max_iter=66)   
```

```python
#return score of successively evaluated points
scores=opt.Y*[-1]
#and plot the best previously observed solution at each iteration
plt.plot([x for x in range(1,len(scores))],[np.max(scores[:i]) for i in range(1,len(scores))])
plt.xlabel("Number of Evaluations")
plt.ylabel("Accuracy")
plt.axhline(y=.68, linewidth=2, color = 'red')
```
